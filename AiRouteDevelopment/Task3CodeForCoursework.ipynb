```python
import pandas as pd
from nbformat import read
import numpy as np
import joblib
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression



# LoadING the provided grid dataset 
# you may need to change this to run the code 
grid_data_path = '/Users/iwuan/OneDrive - hull.ac.uk/_GDrive/Artificial intelligence coursework/AI coursework/provided_grid.csv'
grid_data = pd.read_csv(grid_data_path)

# Display the first few rows to see the structure of the dataset
grid_data.head()
#'/Users/iwuan/OneDrive - hull.ac.uk/_GDrive/Artificial intelligence coursework/AI coursework/task2CodeForCoursework.ipynb' - task2 code file path
```

```python
# Load the trained components from Task 2
model_path = '/Users/iwuan/OneDrive - hull.ac.uk/_GDrive/Artificial intelligence coursework/AI coursework/poly_regression_model.pkl'
scaler_path = '/Users/iwuan/OneDrive - hull.ac.uk/_GDrive/Artificial intelligence coursework/AI coursework/scaler.pkl'
transformer_path = '/Users/iwuan/OneDrive - hull.ac.uk/_GDrive/Artificial intelligence coursework/AI coursework/poly_transformer.pkl'
encoder_path = '/Users/iwuan/OneDrive - hull.ac.uk/_GDrive/Artificial intelligence coursework/AI coursework/encoder.pkl'
```

```python
poly_regression_model = joblib.load(model_path)
scaler = joblib.load(scaler_path)
poly_transformer = joblib.load(transformer_path)
encoder = joblib.load(encoder_path)

# Preprocess the grid data using the loaded encoder and scaler
non_numerical_data = ['type_of_terrain', 'zone_classification', 'time_of_day']

# Apply the encoder to the non-numerical features
encoded_grid_data = pd.DataFrame(
    encoder.transform(grid_data[non_numerical_data]),
    columns=encoder.get_feature_names_out(non_numerical_data)
)

# Drop the original non-numerical features and concatenate the encoded features
numerical_features = grid_data.drop(columns=non_numerical_data)
processed_grid_data = pd.concat([numerical_features, encoded_grid_data], axis=1)

# Apply the scaler to the numerical data
grid_data_scaled = scaler.transform(processed_grid_data)

# Apply the polynomial transformer to generate polynomial features
grid_data_poly = poly_transformer.transform(grid_data_scaled)

# Predict traversal costs using the loaded model
predicted_traversal_costs = poly_regression_model.predict(grid_data_poly)

# Save the predictions to a new DataFrame
processed_grid_data['predicted_traversal_cost'] = predicted_traversal_costs

# Save the results to a CSV file
output_path = '/Users/iwuan/OneDrive - hull.ac.uk/_GDrive/Artificial intelligence coursework/AI coursework/Estimated_grid.csv'
processed_grid_data.to_csv(output_path, index=False)

print(f"Predicted traversal costs saved to: {output_path}")
```
